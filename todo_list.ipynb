```python
# -*- coding: utf-8 -*-
"""
To-Do List (GÃ¶rev YÃ¶netim) UygulamasÄ±
TNC Group - Python ile Temel Programlama Projesi
"""

import os

# Dosya adÄ±
DOSYA_ADI = "gorevler.txt"
```

```python
def gorevleri_yukle():
    """
    GÃ¶revleri dosyadan yÃ¼kler.
    EÄŸer dosya yoksa veya okunamazsa boÅŸ liste dÃ¶ner.
    """
    gorevler = []
    try:
        if os.path.exists(DOSYA_ADI):
            with open(DOSYA_ADI, "r", encoding="utf-8") as dosya:
                gorevler = [satir.strip() for satir in dosya.readlines() if satir.strip()]
            return gorevler, True
        else:
            return gorevler, False
    except Exception as e:
        return gorevler, False
```

```python
def gorevleri_kaydet(gorevler):
    """
    GÃ¶revleri dosyaya kaydeder.
    """
    try:
        with open(DOSYA_ADI, "w", encoding="utf-8") as dosya:
            for gorev in gorevler:
                dosya.write(gorev + "\n")
        return True
    except Exception as e:
        print(f"Dosya yazma hatasÄ±: {e}")
        return False
```

```python
def gorevleri_listele(gorevler):
    """
    GÃ¶revleri numaralandÄ±rÄ±lmÄ±ÅŸ ÅŸekilde listeler.
    """
    if not gorevler:
        print("HenÃ¼z gÃ¶rev bulunmamaktadÄ±r.")
    else:
        print("\n--- GÃ–REV LÄ°STESÄ° ---")
        for i, gorev in enumerate(gorevler, 1):
            print(f"{i}. {gorev}")
```

```python
def yeni_gorev_ekle(gorevler):
    """
    KullanÄ±cÄ±dan yeni gÃ¶rev alÄ±r ve listeye ekler.
    BoÅŸ gÃ¶rev eklenmesine izin verilmez.
    """
    yeni_gorev = input("Yeni gÃ¶rev: ").strip()
    
    if not yeni_gorev:
        print("BoÅŸ gÃ¶rev eklenemez!")
        return False
    
    gorevler.append(yeni_gorev)
    print(f"'{yeni_gorev}' gÃ¶revi eklendi.")
    
    if gorevleri_kaydet(gorevler):
        print("GÃ¶revler baÅŸarÄ±yla kaydedildi.")
        return True
    else:
        return False
```

```python
def gorev_duzenle(gorevler):
    """
    KullanÄ±cÄ±dan gÃ¶rev numarasÄ± alÄ±r ve gÃ¶revi dÃ¼zenler.
    GeÃ§ersiz numara veya boÅŸ gÃ¶rev kontrolÃ¼ yapar.
    """
    if not gorevler:
        print("HenÃ¼z gÃ¶rev bulunmamaktadÄ±r.")
        return False
    
    gorevleri_listele(gorevler)
    
    try:
        gorev_numarasi = input("\nDÃ¼zenlemek istediÄŸiniz gÃ¶revin numarasÄ±: ").strip()
        gorev_numarasi = int(gorev_numarasi)
        
        if gorev_numarasi < 1 or gorev_numarasi > len(gorevler):
            print("GeÃ§ersiz gÃ¶rev numarasÄ±!")
            return False
        
        mevcut_gorev = gorevler[gorev_numarasi - 1]
        yeni_gorev = input(f"Yeni gÃ¶rev metni ({mevcut_gorev}): ").strip()
        
        if not yeni_gorev:
            print("BoÅŸ gÃ¶rev eklenemez!")
            return False
        
        gorevler[gorev_numarasi - 1] = yeni_gorev
        print("GÃ¶rev baÅŸarÄ±yla gÃ¼ncellendi.")
        
        if gorevleri_kaydet(gorevler):
            print("GÃ¶revler baÅŸarÄ±yla kaydedildi.")
            return True
        else:
            return False
            
    except ValueError:
        print("LÃ¼tfen bir sayÄ± girin!")
        return False
    except Exception as e:
        print(f"Hata oluÅŸtu: {e}")
        return False
```

```python
def gorev_sil(gorevler):
    """
    KullanÄ±cÄ±dan gÃ¶rev numarasÄ± alÄ±r ve gÃ¶revi siler.
    GeÃ§ersiz numara kontrolÃ¼ yapar.
    """
    if not gorevler:
        print("HenÃ¼z gÃ¶rev bulunmamaktadÄ±r.")
        return False
    
    gorevleri_listele(gorevler)
    
    try:
        gorev_numarasi = input("\nSilmek istediÄŸiniz gÃ¶revin numarasÄ±: ").strip()
        gorev_numarasi = int(gorev_numarasi)
        
        if gorev_numarasi < 1 or gorev_numarasi > len(gorevler):
            print("GeÃ§ersiz gÃ¶rev numarasÄ±!")
            return False
        
        silinen_gorev = gorevler.pop(gorev_numarasi - 1)
        print(f"'{silinen_gorev}' gÃ¶revi silindi.")
        
        if gorevleri_kaydet(gorevler):
            print("GÃ¶revler baÅŸarÄ±yla kaydedildi.")
            return True
        else:
            return False
            
    except ValueError:
        print("LÃ¼tfen bir sayÄ± girin!")
        return False
    except Exception as e:
        print(f"Hata oluÅŸtu: {e}")
        return False
```

```python
def ana_menu_goster():
    """
    Ana menÃ¼yÃ¼ ekrana yazdÄ±rÄ±r.
    """
    print("\n--- TO-DO LIST UYGULAMASI ---")
    print("1. GÃ¶revleri Listele")
    print("2. Yeni GÃ¶rev Ekle")
    print("3. GÃ¶rev DÃ¼zenle")
    print("4. GÃ¶rev Sil")
    print("5. Ã‡Ä±kÄ±ÅŸ")
```

```python
def main():
    """
    Ana program fonksiyonu.
    """
    print("To-Do List UygulamasÄ±na HoÅŸ Geldiniz!")
    
    # GÃ¶revleri yÃ¼kle
    gorevler, yuklendi = gorevleri_yukle()
    
    if yuklendi:
        print("GÃ¶revler baÅŸarÄ±yla yÃ¼klendi.")
    else:
        print("GÃ¶rev dosyasÄ± bulunamadÄ± veya okunamadÄ±. Yeni bir liste oluÅŸturuldu.")
    
    # Ana dÃ¶ngÃ¼
    while True:
        ana_menu_goster()
        secim = input("\nSeÃ§iminiz (1-5): ").strip()
        
        if secim == "1":
            gorevleri_listele(gorevler)
        elif secim == "2":
            yeni_gorev_ekle(gorevler)
        elif secim == "3":
            gorev_duzenle(gorevler)
        elif secim == "4":
            gorev_sil(gorevler)
        elif secim == "5":
            print("Programdan Ã§Ä±kÄ±lÄ±yor...")
            break
        else:
            print("GeÃ§ersiz seÃ§im! LÃ¼tfen 1-5 arasÄ± bir sayÄ± girin.")

if __name__ == "__main__":
    main()
```
